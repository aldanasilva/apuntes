```python
import seaborn as sns
```

```python
def df_export(dataset, dataset_name: str):
    filepath = f'seaborn/{dataset_name}.parquet'
    dataset.to_parquet(filepath, index=False)
    print(f'Exported: {filepath}')

def df_head(dataset, dataset_name: str):
    print(f'{dataset_name}: {dataset.shape[0]} rows, {dataset.shape[1]} columns.')
    display(dataset.head())

for dataset_name in ['iris', 'planets', 'titanic', 'car_crashes', 'taxis']:
    dataset = sns.load_dataset(dataset_name)
    df_head(dataset, dataset_name)
    df_export(dataset, dataset_name)
    print('\n')
```
